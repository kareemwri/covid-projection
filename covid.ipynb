```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r'D:\Documents\Projects\Covid-19\worldwide.csv')
```

```python
df.head()
```

```python
sns.set(rc={"font.style":"normal",
            "axes.facecolor":(0.9, 0.9, 0.9),
            "figure.facecolor":'white',
            "grid.color":'black',
            "grid.linestyle":':',
            "axes.grid":True,
            'axes.labelsize':20,
            'figure.figsize':(20.0, 10.0),
            'xtick.labelsize':15,
            'ytick.labelsize':15,})
```

```python
d=sns.jointplot(x='Day',y='Deaths', data=df,size=10, order=3, ci=95, scatter_kws={'s':25}, 
                kind='reg',ratio=10,line_kws={"lw":2,'ls':'--'},xlim=(0,100), ylim=(0,200000),color='#d11111')
plt.xlabel('Date')
plt.xticks([0,20,40,60,80,100],['1/22','2/11','3/2','3/22','4/11','5/1'])
plt.legend(['Projected','Actual','Confidence'],borderaxespad=2,loc='upper left',fontsize='large')
plt.title("Projeted Covid-19 Deaths", loc='center', fontsize=20, fontweight=3, color='black',pad=-20)
plt.savefig('deaths.png')
```

```python
c=sns.jointplot(x='Day',y='Confirmed', data=df, size=10, order=2, ci=95, scatter_kws={'s':25}, kind='reg',ratio=10,
                line_kws={"lw":2,'ls':'--'},xlim=(0,100), ylim=(0,1800000),color='#368cba')
plt.xlabel('Date')
plt.xticks([0,20,40,60,80,100],['1/22','2/11','3/2','3/22','4/11','5/1'])
plt.legend(['Projected','Actual','Confidence'],borderaxespad=2,loc='upper left',fontsize='large')
plt.title("Projeted Covid-19 Confirmed Cases", loc='center', fontsize=20, fontweight=3, color='black',pad=-20)
plt.savefig('confirmed.png')
```

```python
r=sns.jointplot(x='Day',y='Recovered', data=df, size=10, order=3, ci=95, scatter_kws={'s':25}, kind='reg',ratio=10,
                line_kws={"lw":2,'ls':'--'},xlim=(0,100), ylim=(0,400000),color='#6ae34f')
plt.xlabel('Date')
plt.xticks([0,20,40,60,80,100],['1/22','2/11','3/2','3/22','4/11','5/1'])
plt.legend(['Projected','Actual','Confidence'],borderaxespad=2,loc='upper left',fontsize='large')
plt.title("Projeted Covid-19 Recoveries", loc='center', fontsize=20, fontweight=3, color='black',pad=-20)
plt.savefig('recovered.png')
```
